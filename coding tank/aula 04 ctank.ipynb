```python
#operadores break e continue

while True:
    cidade = input("digite a cidade escolhida ")
    
    if cidade == 'sair':
        break
    else:    
        print(f'\nVou viajar para {cidade}')
```

```python
adicionais = 0
adicionais_pedido = ""

while True:
    
    if adicionais > 4:
        break
    item = input("digite um adicional: ")
    
    if item == 'sair': 
        break
        
    else:
        adicionais_pedido+=item+" - "
        adicionais+=1
print(adicionais_pedido)
```
