```python
data_atual = time.gmtime()
print(data_atual)
```

```python
ano = data_atual.tm_year
mes = data_atual.tm_mon
dia = data_atual.tm_mday
hora = data_atual.tm_hour
dia_da_semana = data_atual.tm_wday

print('Hoje é dia {}/{}/{}'.format(dia, mes, ano))
```

```python
vendas_meses = [1500, 1727, 1350, 999, 1050, 1027, 1022, 1500, 2000, 2362, 2100, 2762]
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']

import matplotlib.pyplot as plt

plt.plot(meses, vendas_meses)
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.axis([0,12, 0, max(vendas_meses) + 500])
plt.show()
```

```python
import numpy as np
```

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
print(meses)
```

```python
plt.plot(meses, vendas)
plt.axis([0,50,0,max(vendas)+200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
plt.plot(meses, vendas, 'ro')
plt.axis([0,50,0,max(vendas)+200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
plt.plot(meses, vendas, 'r--')
plt.axis([0,50,0,max(vendas)+200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
plt.plot(meses, vendas, 'bs')
plt.axis([0,50,0,max(vendas)+200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
plt.plot(meses, vendas, '^g')
plt.axis([0,50,0,max(vendas)+200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
plt.plot(meses, vendas, color='yellow')
plt.axis([0,50,0,max(vendas)+200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
plt.scatter(meses,vendas)
plt.show()
```

```python
plt.bar(meses, vendas)
plt.show()
```

```python
plt.figure(1, figsize=(15,3))
plt.subplot(1, 3, 1)
plt.plot(meses, vendas, color='yellow')
plt.subplot(1,3,2)
plt.scatter(meses,vendas)
plt.subplot(1,3,3)
plt.bar(meses, vendas)
plt.show()
```
